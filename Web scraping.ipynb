```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"
response=requests.get(url)
print(response.status_code)
```

```python
response.text
```

```python
soup=BeautifulSoup(response.text,"html.parser")  #coverting the raw data if the form of html tree
soup
#indiatable=soup.find('table',{'class': "wikitable"})
```

```python
indiatable=soup.find('table',{'class': "wikitable"})
df=pd.read_html(str(indiatable)) #converting the data fro HTMl to table and creating a list
df=pd.DataFrame(df[0])  #DataFrame combination of row and columns and df is a list having table at the 0th index
df
```

```python
df.to_csv("IndiaTable.csv")
```
